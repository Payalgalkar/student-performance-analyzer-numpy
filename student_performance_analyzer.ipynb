```python
import numpy as np

n = int(input("Enter number of students: "))

marks = np.array([])

for i in range(n):
    mark = int(input(f"Enter marks of student {i+1}: "))
    marks = np.append(marks, mark)

print("This are the marks of students :" ,marks)
print("Total no. of students ",len(marks))#marks.size 
print("Highest marks :" ,np.max(marks))
print("lowest marks :" ,np.min(marks))
print("average marks :" ,np.mean(marks))
print("No. of students who passed :",np.sum(marks>=40))
print("No. of students who failed :",np.sum(marks<40))
print("failed students number :" , marks[marks<40])
print("A Grade students :",np.sum(marks>=80)," ,Students marks are :",marks[marks>80])
print("B Grade students :",np.sum((marks<80) & (marks >=60))," ,Students marks are :",marks[(marks<80) & (marks >=60)])
print("C Grade students :",np.sum((marks<60) & (marks >=40))," ,Students marks are :",marks[(marks<60) & (marks >=40)])
print("D Grade students :",np.sum(marks<40)," ,Students marks are :",marks[marks<40])
print("Topper Marks :", np.max(marks))
avg = np.mean(marks)
print("Above average students marks :",marks[marks>avg])
mar = np.sort(marks)[::-1]
print("Marks Rank")
for  i in mar:
    print(i)
```
